```python
x = list(range(15))
# must change x in equation to another variable 'n' and do 'for n in x' to pull from the x[list]
# y = [n ** 2 for n in x]
y = [n**2 - n+2 for n in x] 

plt.plot(x,y)

plt.annotate('The Origin', xy=(0,0),xytext=(4,125),
             arrowprops={'facecolor': 'blue'})
plt.show()
```
