```python
# Does whether or not a province is Catholic influence fertility? yes
sns.boxplot(x='is_catholic', y='Fertility', data=swiss)
```

```python
from scipy.stats import ttest_ind

catholic_fertility = swiss[swiss['is_catholic'] == True]['Fertility']
non_catholic_fertility = swiss[swiss['is_catholic'] == False]['Fertility']

t_statistic, p_value = ttest_ind(catholic_fertility, non_catholic_fertility)

print('T-Statistic:', t_statistic)
print('P-Value:', p_value)
```

```python
#Load the sleepstudy dataset from pydataset and read it's documentation.
#Use seaborn to create a line chart of all the individual subject's reaction times and a more prominant line 
#showing the average change in reaction time.
```

```python
# Load the sleepstudy dataset
sleepstudy = data('sleepstudy')
```

```python
# Plot a line chart of all subjects' reaction times
sns.lineplot(x='Days', y='Reaction', hue='Subject', data=sleepstudy, alpha=0.4)
```

```python
# Calculate the average change in reaction time over time
mean_reaction = sleepstudy.groupby('Days').mean().reset_index()
mean_reaction.head()
```

```python
# Plot a more prominent line showing the average change in reaction time
sns.lineplot(x='Days', y='Reaction', data=mean_reaction, linewidth=4)
```

```python
# Set up the figure with two subplots
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 4))
sns.lineplot(x='Days', y='Reaction', hue='Subject', data=sleepstudy, alpha=0.4, ax=ax1)
sns.lineplot(x='Days', y='Reaction', data=sleepstudy, estimator='mean', ci=None, linewidth=2, color='black', ax=ax2)
```

```python
# Set up the figure
fig, ax = plt.subplots(figsize=(8, 5))

# Plot a line chart of all subjects' reaction times
for i, subject in enumerate(sleepstudy['Subject'].unique()):
    sns.lineplot(x='Days', y='Reaction', data=sleepstudy[sleepstudy['Subject']==subject],
                 alpha=0.2, ax=ax, label=f'Subject {i+1}')
    
# Plot a more prominent line showing the average change in reaction time
sns.lineplot(x='Days', y='Reaction', data=sleepstudy, estimator='mean', ci=None,
             linewidth=2, color='black', ax=ax, label='Average Change')

# Add labels and title to the plot
ax.set_xlabel('Days')
ax.set_ylabel('Reaction Time')
ax.set_title('Reaction Time by Subject with Average Change')

# Show the legend
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()
```
