```python
# Create the users DataFrame.

users = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['bob', 'joe', 'sally', 'adam', 'jane', 'mike'],
    'role_id': [1, 2, 3, 3, np.nan, np.nan]
})
users
```

```python
# Create the roles DataFrame

roles = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['admin', 'author', 'reviewer', 'commenter']
})
roles
```

```python
# 2. What is the result of using a right join on the DataFrames? lose anything not in right df
```

```python
# In [40]: result = pd.merge(left, right, how="right", on=["key1", "key2"])
pd.merge(roles, users, how="right", left_on='id',right_on='role_id')
```

```python
# Perform an outer join specifying the left and right DataFrame keys.
# Perform an outer join specifying the left and right DataFrame keys.

users.merge(roles, left_on='role_id', right_on='id', how='outer', indicator=True)


```

```python
# 3. What is the result of using an outer join on the DataFrames? brings in the outside info that may not be in both dataframes
```

```python
# 4. What happens if you drop the foreign keys from the DataFrames and try to merge them?
#       #lost its reference points for joint and mixed up the data
pd.merge(users, roles, how='outer')
```
